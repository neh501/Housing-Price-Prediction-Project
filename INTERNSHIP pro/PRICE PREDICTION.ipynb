```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("bangaluruhouseprice.csv")
```

spilling the data into independent and dependent variable

```python
df
```

```python
x=df[["area"]]
```

```python
x
```

```python
y=df.price
```

```python
y
```

data visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.title("area vs price")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(x,y,color="red",s=100,ec=["blue","yellow"])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x
```

```python
model.coef_*5000+model.intercept_
```

```python
model.predict([[5000]])
```

```python
pred=model.predict(x)
```

```python
df
```

```python
plt.title("area vs price")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(x,y,color="red",s=100,ec=["blue","yellow"])
plt.plot(x,pred,color="green")
plt.show()
```
